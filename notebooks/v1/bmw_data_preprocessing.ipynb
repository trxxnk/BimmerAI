```python
import re
import pandas as pd
```

```python
data = pd.read_csv(BMW_PARSED)
data
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
mask = ~data["–û–ø–∏—Å–∞–Ω–∏–µ"].str.contains("—ç–ª–µ–∫—Ç—Ä–æ|–≥–∏–±—Ä–∏–¥", case=False, regex=True, na=False)
data = data[mask].copy()
```

```python
description_columns = ['–ö–æ—Ä–æ–±–∫–∞', '–û–±—ä–µ–º', '–¢–æ–ø–ª–∏–≤–æ', '–ö—É–∑–æ–≤']
data[description_columns] = data['–û–ø–∏—Å–∞–Ω–∏–µ'].str.split(', ', expand=True)
data.drop(columns=['–û–ø–∏—Å–∞–Ω–∏–µ'], inplace=True)
data.head()
```

```python
data['–ú–æ–¥–µ–ª—å'].value_counts()
```

```python
data.groupby('–ú–æ–¥–µ–ª—å').mean(numeric_only=True).sort_values(by='–¶–µ–Ω–∞', ascending=False).round()
```

```python
# –ò–∑–≤–ª–µ–∫–∞–µ–º —Å–µ—Ä–∏—é (–Ω–∞–ø—Ä–∏–º–µ—Ä, "1 —Å–µ—Ä–∏—è", "X6", "7 —Å–µ—Ä–∏—è", "M5")
data["–°–µ—Ä–∏—è"] = data["–ú–æ–¥–µ–ª—å"].str.extract(
    r"BMW (Z\d+|M\d+|[\dX]\s?—Å–µ—Ä–∏—è|X\d+|[\dX]\b)", 
    flags=re.IGNORECASE
)

# –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø–æ–∫–æ–ª–µ–Ω–∏–π (–±—É–∫–≤—ã + —Ü–∏—Ñ—Ä—ã –∏–ª–∏ —Ä–∏–º—Å–∫–∏–µ —Ü–∏—Ñ—Ä—ã)
data["–ü–æ–∫–æ–ª–µ–Ω–∏–µ"] = data["–ú–æ–¥–µ–ª—å"].str.findall(
    r"([A-Z]\d{2,3}|[IVXLCDM]+)\b"
).apply(lambda x: ', '.join(x) if x else None)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ä–µ—Å—Ç–∞–π–ª–∏–Ω–≥
data["–†–µ—Å—Ç–∞–π–ª–∏–Ω–≥"] = data["–ú–æ–¥–µ–ª—å"].str.contains("–†–µ—Å—Ç–∞–π–ª–∏–Ω–≥", case=False).astype(int)
```
