```python
import re
import pandas as pd
```

```python
data = pd.read_csv("../data/bmw_data.csv")
data
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
mask = ~data["Описание"].str.contains("электро|гибрид", case=False, regex=True, na=False)
data = data[mask].copy()
```

```python
description_columns = ['Коробка', 'Объем', 'Топливо', 'Кузов']
data[description_columns] = data['Описание'].str.split(', ', expand=True)
data.drop(columns=['Описание'], inplace=True)
data.head()
```

```python
data['Модель'].value_counts()
```

```python
data.groupby('Модель').mean(numeric_only=True).sort_values(by='Цена', ascending=False).round()
```

```python
# Извлекаем серию (например, "1 серия", "X6", "7 серия", "M5")
data["Серия"] = data["Модель"].str.extract(
    r"BMW (Z\d+|M\d+|[\dX]\s?серия|X\d+|[\dX]\b)", 
    flags=re.IGNORECASE
)

# Извлечение поколений (буквы + цифры или римские цифры)
data["Поколение"] = data["Модель"].str.findall(
    r"([A-Z]\d{2,3}|[IVXLCDM]+)\b"
).apply(lambda x: ', '.join(x) if x else None)

# Проверка на рестайлинг
data["Рестайлинг"] = data["Модель"].str.contains("Рестайлинг", case=False).astype(int)
```

```python
data[['Серия', 'Поколение']].value_counts().sort_values(ascending=False).to_frame()
```

```python
data.drop(["Модель"], axis=1, inplace=True, errors="ignore")
data
```

```python
description_columns = [
    "Серия","Поколение","Кузов","Год","Рестайлинг","Пробег","Коробка","Объем","Топливо","Цена"
]
data = data[description_columns]
data
```
